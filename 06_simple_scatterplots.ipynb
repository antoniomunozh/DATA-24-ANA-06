```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
FILENAME = 'tmp/Ejemplo_02_03_Cybersecurity_Dataset.csv'
```

```python
# Cargar nuestro dataset
df = pd.read_csv(FILENAME)
df.head(10)
```

```python
sns.set(style='darkgrid')
# Configurar scatter
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='severity', y='response_time', color='b',s=100)
# Cofigurar titulo y etiquetas
plt.title('Severidad VS tiempo de respuesta')
plt.xlabel('Severidad del incidente')
plt.ylabel('Tiempo de respuesta (Horas)')
plt.show()
```
